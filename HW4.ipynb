# Part A: Multi-Document Parser (20 points)

## What this part is about

Before the rule engine can evaluate anything, your system needs to understand the structure of what it received. In the real world, a legal AI system does not receive a single clean contract, it receives a package of documents, and those documents interact with each other in ways that matter enormously for risk assessment.

This part builds two deterministic (no LLM) components. The first is a **document classifier** that reads a document and identifies what type it is: an original agreement, an amendment that modifies an original, or a schedule that adds scope and special terms. Classification also extracts which clause types the document covers and whether the document makes any explicit claim about controlling other documents.

The second is a **conflict detector**. When two documents both address the same clause, say both have payment terms, your detector compares the values and determines which document controls. This is a non-trivial problem because the documents in this package make conflicting precedence claims. The Schedule says it controls over the MSA and any Amendment on payment and IP matters. The Amendment says it supersedes the MSA on payment terms. When both documents claim control over the same clause, the right answer is not to pick one arbitrarily, it is to flag the ambiguity with low confidence and explain why.

## The three conflicts you must find

Read the document package carefully and identify these before coding:

1. **Payment terms** appear in all three documents with different values (60, 90, and 120 days), and both the Amendment and the Schedule claim their value controls. This is genuinely ambiguous. Your detector must flag it as low-confidence resolution.

2. **IP ownership** is granted to the Client by the MSA, then partially reversed by the Schedule for ML models trained on Client data. A system that only reads the MSA will miss this entirely.

3. **Termination rights** are symmetric in the MSA (neither party can terminate for convenience), but the Amendment gives the Vendor unilateral 7-day termination. This is a significant power shift that only becomes visible when you compare the two documents.

## Why the homework uses deterministic classification

As noted in the architecture slide, document classification is genuinely a neural task in production. A fine-tuned model handles the linguistic variation that keyword rules cannot. But in this homework you will implement the classifier with keyword heuristics, and the reason is architectural rather than a statement about what is technically possible.

The classifier feeds the conflict detector, which feeds the rule engine, which produces findings that must be fully auditable. A stochastic step anywhere in that chain means two runs on the same document could produce different risk reports, and you cannot explain which one is correct. For this homework, deterministic classification keeps that audit trail clean and lets you focus on the harder design challenges in the layers above it. In a production system, the equivalent guarantee comes from confidence thresholds and human review for low-confidence predictions, not from removing the model entirely.

**Grading rubric:** `DocumentClassifier` correctly identifies all three types (5 pts). `ConflictDetector` finds all three conflicts (8 pts). Precedence logic is internally consistent and handles the ambiguous three-way payment conflict with `LOW` confidence (7 pts).

````python
# Part A: Step 1: Document Classifier

from dataclasses import dataclass, field
from typing import Optional
import re

@dataclass
class DocumentMeta:
    doc_id: str
    doc_type: str                  # 'MSA', 'Amendment', 'Schedule', 'Unknown'
    effective_date: Optional[str]  # date string as it appears in the document
    parties: list                  # extracted party names
    precedence_claims: list        # full sentences claiming this doc controls others
    clauses_covered: list          # which clause types this document addresses


class DocumentClassifier:

    # Keywords that signal each document TYPE
    TYPE_KEYWORDS = {
        'Amendment': [
            'amendment', 'amended section', 'supersedes', 'replaces',
            'deleted in its entirety', 'is hereby removed', 'effective immediately'
        ],
        'Schedule': [
            'schedule', 'exhibit', 'addendum', 'scope of services',
            'services in scope', 'phase 1', 'phase 2', 'phase 3'
        ],
        'MSA': [
            'master services agreement', 'msa', 'agrees to provide',
            'limitation of liability', 'indemnification', 'governing law'
        ],
    }

    # Keywords that signal each CLAUSE TYPE is present in a document
    CLAUSE_KEYWORDS = {
        'payment_terms':      ['payment', 'invoice', 'net-30', 'days of receipt', 'late payment'],
        'ip_ownership':       ['intellectual property', 'work product', 'deliverable', 'ownership',
                               'exclusive property', 'ml model', 'trained on'],
        'governing_law':      ['governing law', 'governed by', 'laws of the state', 'jurisdiction'],
        'confidentiality':    ['confidential', 'confidentiality', 'hold in confidence'],
        'termination':        ['termination', 'terminate', 'non-renewal', 'notice of non-renewal'],
        'indemnification':    ['indemnif'],
        'liability_cap':      ['limitation of liability', 'liability shall not exceed',
                               'total liability', 'liability cap'],
        'dispute_resolution': ['arbitration', 'dispute', 'jams', 'mediation'],
        'non_compete':        ['non-compete', 'noncompete', 'compete', 'competitive'],
        'subcontractor':      ['subcontractor', 'subcontract', 'third-party provider'],
    }

    # Sentences containing these phrases are precedence claims
    PRECEDENCE_PHRASES = [
        'shall control', 'notwithstanding', 'supersedes', 'replaces',
        'deleted in its entirety', 'is hereby removed', 'takes precedence',
        'governs over'
    ]

    def classify(self, doc_id: str, text: str) -> DocumentMeta:
        text_lower = text.lower()

        doc_type   = self._detect_type(text_lower)
        date       = self._extract_date(text)
        parties    = self._extract_parties(text)
        precedence = self._extract_precedence_claims(text)
        clauses    = self._detect_clauses(text_lower)

        return DocumentMeta(
            doc_id            = doc_id,
            doc_type          = doc_type,
            effective_date    = date,
            parties           = parties,
            precedence_claims = precedence,
            clauses_covered   = clauses,
        )

    def _detect_type(self, text_lower: str) -> str:
          # Check the title lines first — most reliable signal
          first_lines = '\n'.join(text_lower.strip().splitlines()[:3])

          if 'amendment' in first_lines:
              return 'Amendment'
          if 'schedule' in first_lines:
              return 'Schedule'
          if 'master services agreement' in first_lines:
              return 'MSA'

          # Fallback to body keywords
          if any(kw in text_lower for kw in self.TYPE_KEYWORDS['Amendment']):
              return 'Amendment'
          if any(kw in text_lower for kw in self.TYPE_KEYWORDS['Schedule']):
              return 'Schedule'
          if any(kw in text_lower for kw in self.TYPE_KEYWORDS['MSA']):
              return 'MSA'

          return 'Unknown'

    def _extract_date(self, text: str) -> Optional[str]:
        # Match patterns like "January 15, 2026" or "March 1, 2026"
        match = re.search(
            r'(?:as of|entered into|effective)\s+([A-Z][a-z]+ \d{1,2},\s*\d{4})',
            text
        )
        return match.group(1) if match else None

    def _extract_parties(self, text: str) -> list:
        # Look for quoted party names followed by role labels
        parties = re.findall(
            r'"([^"]+?)"\s*\(\s*"(?:Client|Vendor|Company|Contractor|Party)"\s*\)',
            text
        )
        return list(dict.fromkeys(parties))  # deduplicate, preserve order

    def _extract_precedence_claims(self, text: str) -> list:
        claims = []
        for sentence in re.split(r'(?<=[.!?])\s+', text):
            sentence_lower = sentence.lower()
            if any(phrase in sentence_lower for phrase in self.PRECEDENCE_PHRASES):
                claims.append(sentence.strip())
        return claims

    def _detect_clauses(self, text_lower: str) -> list:
        found = []
        for clause, keywords in self.CLAUSE_KEYWORDS.items():
            if any(kw in text_lower for kw in keywords):
                found.append(clause)
        return found
````

````python
# Part A: Step 2: Conflict Detector

from dataclasses import dataclass
from typing import Any

@dataclass
class ConflictReport:
    conflicts: list        # list of conflict dicts
    has_conflicts: bool
    resolution_confidence: str   # 'HIGH', 'MEDIUM', or 'LOW'


class ConflictDetector:

    # Which fields to compare between documents, and which clause they belong to
    FIELD_TO_CLAUSE = {
        'payment_terms_days':           'payment_terms',
        'ip_ownership':                 'ip_ownership',
        'ip_covers_trained_models':     'ip_ownership',
        'termination_for_convenience':  'termination',
        'termination_notice_days':      'termination',
        'governing_law':                'governing_law',
        'liability_cap_present':        'liability_cap',
        'non_compete_years':            'non_compete',
    }

    def detect(self, meta_a: DocumentMeta, fields_a: dict,
               meta_b: DocumentMeta, fields_b: dict) -> ConflictReport:

        conflicts = []

        for field_name, clause in self.FIELD_TO_CLAUSE.items():
            val_a = fields_a.get(field_name)
            val_b = fields_b.get(field_name)

            # Only a conflict if both docs have a value AND they differ
            if val_a is None or val_b is None:
                continue
            if val_a == val_b:
                continue

            controlling_doc, controlling_val, rationale, confidence = \
                self._resolve(field_name, clause, meta_a, val_a, meta_b, val_b)

            # Risk is amplified if the controlling value is worse for the Client
            risk_amplified = self._is_risk_amplified(field_name, controlling_val,
                                                      val_a, val_b, meta_a, meta_b)

            conflicts.append({
                'clause':            clause,
                'field':             field_name,
                'doc_a':             meta_a.doc_id,
                'doc_a_value':       val_a,
                'doc_b':             meta_b.doc_id,
                'doc_b_value':       val_b,
                'controlling_doc':   controlling_doc,
                'controlling_value': controlling_val,
                'rationale':         rationale,
                'risk_amplified':    risk_amplified,
                'confidence':        confidence,
            })

        overall_confidence = self._overall_confidence(conflicts)

        return ConflictReport(
            conflicts             = conflicts,
            has_conflicts         = len(conflicts) > 0,
            resolution_confidence = overall_confidence,
        )

    def _resolve(self, field_name, clause, meta_a, val_a, meta_b, val_b):
        """
        Apply precedence rules in order:
          1. Both docs claim control over this clause → LOW confidence
          2. Amendment explicitly supersedes MSA for this clause → Amendment wins
          3. Schedule claims notwithstanding control → Schedule wins
          4. Default: later document wins (doc with higher doc_id number)
        """
        a_claims = self._claims_control(meta_a, clause)
        b_claims = self._claims_control(meta_b, clause)

        # Rule 1: Both claim control — genuinely ambiguous
        if a_claims and b_claims:
            return (
                meta_b.doc_id,   # tentatively pick the later doc
                val_b,
                (f"AMBIGUOUS: Both {meta_a.doc_id} ({meta_a.doc_type}) and "
                 f"{meta_b.doc_id} ({meta_b.doc_type}) claim precedence over "
                 f"'{clause}'. {meta_a.doc_id} value={val_a}, "
                 f"{meta_b.doc_id} value={val_b}. "
                 f"Legal review required to determine which controls."),
                'LOW'
            )

        # Rule 2: One side is an Amendment that explicitly replaces this clause
        if meta_a.doc_type == 'Amendment' and clause in meta_a.clauses_covered:
            return (meta_a.doc_id, val_a,
                    f"Amendment {meta_a.doc_id} explicitly replaces '{clause}' in the base agreement.",
                    'HIGH')
        if meta_b.doc_type == 'Amendment' and clause in meta_b.clauses_covered:
            return (meta_b.doc_id, val_b,
                    f"Amendment {meta_b.doc_id} explicitly replaces '{clause}' in the base agreement.",
                    'HIGH')

        # Rule 3: One side is a Schedule with a notwithstanding claim
        if meta_a.doc_type == 'Schedule' and a_claims:
            return (meta_a.doc_id, val_a,
                    f"Schedule {meta_a.doc_id} contains a 'notwithstanding' claim over '{clause}'.",
                    'MEDIUM')
        if meta_b.doc_type == 'Schedule' and b_claims:
            return (meta_b.doc_id, val_b,
                    f"Schedule {meta_b.doc_id} contains a 'notwithstanding' claim over '{clause}'.",
                    'MEDIUM')

        # Rule 4: Default — later document wins
        later = meta_b if meta_b.doc_id > meta_a.doc_id else meta_a
        later_val = val_b if later == meta_b else val_a
        return (later.doc_id, later_val,
                f"No explicit precedence claim. Later document ({later.doc_id}) controls by default.",
                'MEDIUM')

    def _claims_control(self, meta: DocumentMeta, clause: str) -> bool:
        """
        Returns True if this document has a precedence claim
        that is plausibly about the given clause.
        """
        clause_keywords = {
            'payment_terms': ['payment', 'invoice'],
            'ip_ownership':  ['ip', 'intellectual property', 'model', 'deliverable'],
            'termination':   ['terminat'],
            'governing_law': ['governing', 'jurisdiction'],
        }
        keywords = clause_keywords.get(clause, [])
        for claim in meta.precedence_claims:
            claim_lower = claim.lower()
            if any(kw in claim_lower for kw in keywords):
                return True
        return False

    def _is_risk_amplified(self, field_name, controlling_val,
                            val_a, val_b, meta_a, meta_b) -> bool:
        """
        Returns True if the controlling value is worse for the Client
        than the alternative value would have been.
        """
        # Longer payment terms = more risk for Client (cash flow)
        if field_name == 'payment_terms_days':
            other_val = val_a if controlling_val == val_b else val_b
            return (isinstance(controlling_val, (int, float)) and
                    isinstance(other_val, (int, float)) and
                    controlling_val > other_val)

        # Vendor owning IP = worse for Client
        if field_name in ('ip_ownership', 'ip_covers_trained_models'):
            return str(controlling_val).lower() in ('vendor', 'true', True)

        # Vendor-only termination = worse for Client
        if field_name == 'termination_for_convenience':
            return str(controlling_val).lower() == 'vendor_only'

        return False

    def _overall_confidence(self, conflicts: list) -> str:
        if not conflicts:
            return 'HIGH'
        confidences = [c['confidence'] for c in conflicts]
        if 'LOW' in confidences:
            return 'LOW'
        if 'MEDIUM' in confidences:
            return 'MEDIUM'
        return 'HIGH'
````

# Part B: Jurisdiction-Aware Symbolic Rule Engine (25 points)

## What this part is about

The rule engine is the core of the symbolic component in your neuro-symbolic system. It takes a structured record of contract field values and deterministically computes a risk assessment. The same input always produces the same output. There is no LLM involved. A lawyer who reads your rule code should be able to verify each decision without any knowledge of machine learning.

This part introduces two ideas that go beyond a simple flat rule list.

The first is **jurisdiction-aware thresholds**. Contract risk does not exist in a vacuum; it exists relative to the law that governs the contract. A non-compete clause in Texas may be enforceable with reasonable scope. The same clause in California is void by statute (California Business and Professions Code Section 16600 makes most non-competes unenforceable). A rule engine that does not account for jurisdiction will either over-flag risks in Texas or under-flag them in California. Your engine must load the correct threshold set once governing law is known.

The second is **compositional (Tier 2) rules**. Some risks are only visible when multiple clause-level findings are combined. Imagine a contract where the Vendor can terminate with 7 days notice, the indemnification runs only one way against the Client, and the ML models trained on Client data belong to the Vendor. Any one of these clauses, reviewed in isolation, produces a HIGH finding. But the combination is qualitatively worse, it means the Vendor can build commercial products from the Client's data, disclaim liability, and walk away in a week. That is a pattern, not just a collection of individual flags. Tier 2 rules detect these patterns.

## Why the rule engine must be stateful across documents

In the previous version of this homework, the rule engine was a stateless function: give it a `ContractFields` object, get a report back. That works for a single document. Here, the Tier 2 compositional rules need to see findings from the merged, conflict-resolved field set, the combined picture of all three documents. The engine accumulates findings before it can fire pattern rules.

## Jurisdiction background you need to know

You do not need to be a lawyer to implement this, but you do need to understand the legal basis for the threshold differences:

**California payment terms:** The California Prompt Payment Act establishes 30-day payment defaults for many service contracts. A 60-day term is already above the statutory default, making it MEDIUM risk. 75+ days is HIGH.

**California non-competes:** Section 16600 of the California Business and Professions Code renders most employee and contractor non-compete clauses void and unenforceable. Any non-compete in a California-governed contract is HIGH risk regardless of its duration or geographic scope, not because it is enforceable, but because including it may expose the Vendor to legal liability for attempting to enforce an illegal clause.

**California ML model IP:** California's consumer privacy protections (CCPA) create additional obligations when a Vendor uses Client data to train commercial models. Even if the IP clause is contractually valid, the data use may trigger regulatory requirements that the contract does not address.

**Grading rubric:** `ContractFields` with enhanced fields (3 pts). Tier 1 rules with jurisdiction-specific thresholds for at least 3 rules (10 pts). `JurisdictionRuleSetFactory` produces correct thresholds for California, Texas, and default (5 pts). Tier 2 patterns with at least 3 compositional rules (5 pts). Unit tests pass (2 pts).

````python
# Part B: Step 1: Enhanced ContractFields

from dataclasses import dataclass, field
from typing import Optional

@dataclass
class ContractFields:
    # --- Original fields ---
    governing_law:                  Optional[str]   = None  # e.g. 'California', 'Texas'
    payment_terms_days:             Optional[int]   = None  # e.g. 60, 90, 120
    liability_cap_present:          Optional[bool]  = None
    liability_cap_basis:            Optional[str]   = None  # e.g. 'fees_paid', 'fixed'
    ip_ownership:                   Optional[str]   = None  # 'client' or 'vendor'
    auto_renewal:                   Optional[bool]  = None
    auto_renewal_notice_days:       Optional[int]   = None
    termination_for_convenience:    Optional[str]   = None  # 'none', 'vendor_only', 'both'
    termination_notice_days:        Optional[int]   = None  # legacy single field
    indemnification_direction:      Optional[str]   = None  # 'mutual', 'client_only', 'vendor_only'
    non_compete_years:              Optional[float] = None
    non_compete_scope:              Optional[str]   = None  # e.g. 'regional', 'global'
    subcontractor_approval_required: Optional[bool] = None
    confidentiality_survival_years: Optional[float] = None
    dispute_resolution:             Optional[str]   = None  # 'arbitration', 'litigation'

    # --- New fields ---
    controlling_payment_terms_days: Optional[int]   = None
    # Payment term after conflict resolution. May differ from payment_terms_days
    # if a different document was found to control.

    controlling_document_id:        Optional[str]   = None
    # Which document's value is being used as the controlling value.
    # Used in the audit trail so findings can be traced to a source document.

    has_unresolved_conflict:        bool            = False
    # True if any conflict was flagged with LOW resolution confidence.
    # The rule engine fires R12 on this field.

    ip_covers_trained_models:       bool            = False
    # True if the IP clause assigns ownership of ML models trained on
    # Client data to the Vendor.

    termination_notice_vendor_days: Optional[int]   = None
    # Vendor-side notice period for termination.
    # In the Amendment, vendor gets 7 days.

    termination_notice_client_days: Optional[int]   = None
    # Client-side notice period for termination.
    # In the Amendment, client gets 60 days.
    # A single termination_notice_days field loses this asymmetry.
````

````python
# Part B: Step 2: JurisdictionRuleSet and Factory

from dataclasses import dataclass
from typing import Optional

@dataclass
class JurisdictionRuleSet:
    jurisdiction:                   str
    payment_medium_threshold_days:  int
    payment_high_threshold_days:    int
    non_compete_base_level:         str           # 'LOW', 'MEDIUM', or 'HIGH'
    trained_model_ip_extra_flag:    Optional[str] # None or a CCPA warning string


class JurisdictionRuleSetFactory:

    @classmethod
    def get(cls, governing_law: Optional[str]) -> JurisdictionRuleSet:
        if governing_law and 'california' in governing_law.lower():
            return JurisdictionRuleSet(
                jurisdiction                  = 'California',
                payment_medium_threshold_days = 45,
                payment_high_threshold_days   = 75,
                non_compete_base_level        = 'HIGH',
                trained_model_ip_extra_flag   = (
                    'CCPA risk: Vendor use of Client data in commercial models '
                    'may require Client consent under Cal. Civ. Code §1798.100'
                ),
            )

        if governing_law and 'texas' in governing_law.lower():
            return JurisdictionRuleSet(
                jurisdiction                  = 'Texas',
                payment_medium_threshold_days = 60,
                payment_high_threshold_days   = 90,
                non_compete_base_level        = 'MEDIUM',
                trained_model_ip_extra_flag   = None,
            )

        # Default — unknown or any other jurisdiction
        return JurisdictionRuleSet(
            jurisdiction                  = governing_law or 'Unknown',
            payment_medium_threshold_days = 60,
            payment_high_threshold_days   = 90,
            non_compete_base_level        = 'MEDIUM',
            trained_model_ip_extra_flag   = None,
        )
````

````python
# Part B: Step 3: Two-Tier Rule Engine

class ContractRuleEngine:

    RULE_VERSION = '1.0.0'

    def __init__(self, ruleset: JurisdictionRuleSet):
        self.ruleset = ruleset

    def generate_report(self, fields: ContractFields) -> dict:
        tier1 = self._run_tier1(fields)
        tier2 = self._run_tier2(fields, tier1)

        fired_levels = [f['level'] for f in tier1 + tier2]
        if 'CRITICAL' in fired_levels:
            overall = 'CRITICAL'
        elif 'HIGH' in fired_levels:
            overall = 'HIGH'
        elif 'MEDIUM' in fired_levels:
            overall = 'MEDIUM'
        else:
            overall = 'LOW'

        return {
            'overall_risk':    overall,
            'tier1_findings':  tier1,
            'tier2_findings':  tier2,
            'jurisdiction':    self.ruleset.jurisdiction,
            'critical_count':  fired_levels.count('CRITICAL'),
            'high_count':      fired_levels.count('HIGH'),
            'medium_count':    fired_levels.count('MEDIUM'),
            'rule_version':    self.RULE_VERSION,
        }

    # ------------------------------------------------------------------
    # TIER 1 RULES
    # ------------------------------------------------------------------

    def _run_tier1(self, f: ContractFields) -> list:
        findings = []
        for rule in [
            self._r01, self._r02, self._r03, self._r04,
            self._r05, self._r06, self._r07, self._r08,
            self._r09, self._r10, self._r11, self._r12,
        ]:
            result = rule(f)
            if result:
                findings.append(result)
        return findings

    def _r01(self, f: ContractFields):
        days = f.controlling_payment_terms_days or f.payment_terms_days
        if days is None:
            return None
        if days > self.ruleset.payment_high_threshold_days:
            level = 'HIGH'
        elif days > self.ruleset.payment_medium_threshold_days:
            level = 'MEDIUM'
        else:
            return None
        return {
            'field': 'R01',
            'level': level,
            'explanation': (
                f"Payment terms of {days} days exceed the {self.ruleset.jurisdiction} "
                f"{level.lower()} threshold of {self.ruleset.payment_high_threshold_days if level == 'HIGH' else self.ruleset.payment_medium_threshold_days} days, "
                f"creating cash flow risk and potential suspension rights for the Vendor."
            )
        }

    def _r02(self, f: ContractFields):
        if f.liability_cap_present is False:
            return {
                'field': 'R01',
                'level': 'HIGH',
                'explanation': (
                    "No liability cap is present. Without a cap, the Client's exposure "
                    "to damages is unlimited, which is standard practice to limit."
                )
            }
        if f.liability_cap_present and f.liability_cap_basis not in ('fees_paid', 'fixed'):
            return {
                'field': 'R02',
                'level': 'MEDIUM',
                'explanation': (
                    f"Liability cap basis is '{f.liability_cap_basis}', which is vague "
                    f"and may be difficult to enforce or calculate in a dispute."
                )
            }
        return None

    def _r03(self, f: ContractFields):
        if f.ip_ownership == 'vendor':
            return {
                'field': 'R03',
                'level': 'HIGH',
                'explanation': (
                    "IP ownership of custom deliverables is assigned to the Vendor rather "
                    "than the Client. The Client pays for work it does not own."
                )
            }
        return None

    def _r04(self, f: ContractFields):
        if not f.ip_covers_trained_models:
            return None
        explanation = (
            "The IP clause assigns ownership of ML models trained on Client data to "
            "the Vendor. These models may encode proprietary Client insights and "
            "be incorporated into Vendor commercial products."
        )
        if self.ruleset.trained_model_ip_extra_flag:
            explanation += f" {self.ruleset.trained_model_ip_extra_flag}"
        return {
            'field': 'R04',
            'level': 'HIGH',
            'explanation': explanation,
        }

    def _r05(self, f: ContractFields):
        if f.auto_renewal and f.auto_renewal_notice_days is not None:
            if f.auto_renewal_notice_days < 14:
                return {
                    'field': 'R05',
                    'level': 'MEDIUM',
                    'explanation': (
                        f"Auto-renewal notice period is only {f.auto_renewal_notice_days} days, "
                        f"which is below the 14-day minimum recommended to give the Client "
                        f"adequate time to evaluate renewal."
                    )
                }
        return None

    def _r06(self, f: ContractFields):
        if f.termination_for_convenience == 'vendor_only':
            return {
                'field': 'R06',
                'level': 'HIGH',
                'explanation': (
                    "The Vendor may terminate this agreement for any reason with short notice. "
                    "The Client has no equivalent right. This creates a significant power imbalance."
                )
            }
        return None

    def _r07(self, f: ContractFields):
        vendor_days = f.termination_notice_vendor_days
        client_days = f.termination_notice_client_days
        if vendor_days is not None and client_days is not None:
            if vendor_days < 14 and client_days > 30:
                return {
                    'field': 'R07',
                    'level': 'HIGH',
                    'explanation': (
                        f"Termination notice is severely asymmetric: Vendor needs only "
                        f"{vendor_days} days to exit, while Client requires {client_days} days "
                        f"to terminate for breach. The Client is effectively locked in."
                    )
                }
        return None

    def _r08(self, f: ContractFields):
        if f.indemnification_direction == 'client_only':
            return {
                'field': 'R08',
                'level': 'HIGH',
                'explanation': (
                    "Indemnification runs only against the Client. The Vendor bears no "
                    "indemnification obligations, leaving the Client exposed to all "
                    "third-party claims."
                )
            }
        return None

    def _r09(self, f: ContractFields):
        if f.non_compete_years is None:
            return None
        level = self.ruleset.non_compete_base_level
        basis = (
            "California Business & Professions Code §16600 renders most non-competes void."
            if self.ruleset.jurisdiction == 'California'
            else "Non-compete clauses are enforceable if reasonable in scope and duration."
        )
        return {
            'field': 'R09',
            'level': level,
            'explanation': (
                f"A non-compete clause of {f.non_compete_years} year(s) "
                f"({f.non_compete_scope or 'unspecified'} scope) is present. {basis}"
            )
        }

    def _r10(self, f: ContractFields):
        if f.subcontractor_approval_required is False:
            return {
                'field': 'R10',
                'level': 'MEDIUM',
                'explanation': (
                    "The Vendor may engage subcontractors without Client approval. "
                    "The Client has no visibility into who handles its data or deliverables."
                )
            }
        return None

    def _r11(self, f: ContractFields):
        if f.governing_law is None:
            return {
                'field': 'R11',
                'level': 'MEDIUM',
                'explanation': (
                    "No governing law is specified in any document. This creates uncertainty "
                    "about which jurisdiction's rules apply to disputes and risk thresholds."
                )
            }
        return None

    def _r12(self, f: ContractFields):
        if f.has_unresolved_conflict:
            return {
                'field': 'R12',
                'level': 'MEDIUM',
                'explanation': (
                    "At least one cross-document conflict could not be resolved with high "
                    "confidence. Ambiguous precedence creates legal uncertainty — a court "
                    "may apply a different value than the one your system selected."
                )
            }
        return None

    # ------------------------------------------------------------------
    # TIER 2 COMPOSITIONAL RULES
    # ------------------------------------------------------------------

    def _run_tier2(self, f: ContractFields, tier1: list) -> list:
        fired = {r['field'] for r in tier1}
        findings = []
        for rule in [self._c01, self._c02, self._c03]:
            result = rule(f, fired)
            if result:
                findings.append(result)
        return findings

    def _c01(self, f: ContractFields, fired: set):
        # Vendor termination + one-sided indemnity + IP to vendor
        if 'R06' in fired and 'R08' in fired and ('R03' in fired or 'R04' in fired):
            return {
                'field': 'C01',
                'level': 'CRITICAL',
                'explanation': (
                    "STRUCTURAL IMBALANCE: The Vendor can exit in days (R06), "
                    "bears no indemnification obligations (R08), and owns the work "
                    "product or Client-trained models (R03/R04). The Client has all "
                    "the obligations and none of the protections. This combination is "
                    "qualitatively worse than any single clause in isolation."
                )
            }
        return None

    def _c02(self, f: ContractFields, fired: set):
        # ML model IP to vendor + subcontractor without approval
        if 'R04' in fired and 'R10' in fired:
            return {
                'field': 'C02',
                'level': 'CRITICAL',
                'explanation': (
                    "DATA SOVEREIGNTY RISK: The Vendor owns ML models trained on "
                    "Client data (R04) and can share the pipeline with unapproved "
                    "subcontractors (R10). The Client has no visibility into where "
                    "its data ends up or who ultimately controls those models."
                )
            }
        return None

    def _c03(self, f: ContractFields, fired: set):
        # High payment terms + vendor short-notice termination
        r01 = next((r for r in fired if r == 'R01'), None)
        if r01 and 'R06' in fired:
            return {
                'field': 'C03',
                'level': 'CRITICAL',
                'explanation': (
                    "PAYMENT AND TERMINATION TRAP: The Client owes extended payment "
                    "after milestone completion (R01), but the Vendor can terminate "
                    "in as few as 7 days at any point during that window (R06). "
                    "The Client could be in default on payment for work that was "
                    "terminated before it could be completed."
                )
            }
        return None
````

# Part C: Neural Extraction with Grounding Verification (15 points)

## What this part is about

The neural extractor is where the LLM does its legitimate work: reading raw contract text and populating structured fields that the rule engine can evaluate. This is the only component in your system that processes unstructured natural language. Everything downstream of it, conflict detection, rule evaluation, agent reasoning, works with structured data.

The central challenge in Part C is **grounding verification**. An LLM asked to extract a payment term from a minimal one-paragraph contract may return 30 as a default because that is common in training data, even if the document says nothing about payment terms at all. This is not a bug in the model, it is doing exactly what it was trained to do: produce a fluent, plausible response. But in a legal system, a hallucinated field value is a serious failure. It can cause the rule engine to miss a risk (if the LLM fills in a safe-looking value for a missing field) or to flag a false risk (if the LLM invents a risky value).

Grounding verification is your defense against this. After the LLM extracts a value for a field, you check whether that value can be traced back to actual text in the document. If it cannot, the field is marked as unverified and treated as absent by the rule engine. The check must be deterministic, no second LLM call, no fuzzy matching library, just string operations on the original document text.

## Document-type-aware extraction

The extractor must handle documents differently based on their type. An MSA is a complete contract: every field either exists or is absent. An Amendment is a delta; it only changes specific fields, and all other fields remain as they were in the MSA. If you ask an LLM to extract all fields from an Amendment and it returns a value for `liability_cap_present`, that value is almost certainly hallucinated, the Amendment says nothing about liability.

Your prompt for Amendment extraction must explicitly instruct the LLM to return null for any field not directly addressed. The `merge_fields` function then applies Amendment values on top of MSA values, only overwriting fields that the Amendment actually changed.

## What grounding means operationally

For a numeric field like `payment_terms_days`, grounding means the exact number appears as a digit string in the document. If the LLM extracted 90 and the document contains the string "90" or "ninety (90)", the field is grounded. If the document does not contain either, it is not grounded.

For an enum field like `ip_ownership`, grounding means at least one keyword from a predefined keyword map appears near a relevant context phrase. You define the keyword map. For `ip_ownership = 'vendor'`, the keywords might be `['shall remain the property of vendor', 'exclusive property of vendor', 'vendor retains']`. The document must contain at least one of these to ground the extracted value.

For a boolean field like `subcontractor_approval_required`, grounding means the document contains a relevant phrase about subcontractors. You define what counts as relevant.

**Grading rubric:** Document-type-aware extraction handles both MSA and Amendment correctly (6 pts). Grounding verification catches at least one hallucination in the adversarial test (5 pts). `merge_fields` applies Amendment deltas correctly and logs all changed fields (4 pts).

````python
# Part C: Step 1: LLM Setup

import os
from groq import Groq
from langchain_core.language_models.llms import LLM
from langchain_core.callbacks.manager import CallbackManagerForLLMRun
from typing import Any, List, Optional

GROQ_API_KEY = 'groq'

class GroqLLM(LLM):
    model: str = 'llama-3.3-70b-versatile'
    api_key: str = ''

    def __init__(self, model: str, api_key: str = '', **kwargs):
        super().__init__(model=model, api_key=api_key, **kwargs)

    @property
    def _llm_type(self) -> str:
        return 'groq'

    def _call(
        self,
        prompt: str,
        stop: Optional[List[str]] = None,
        run_manager: Optional[CallbackManagerForLLMRun] = None,
        **kwargs: Any,
    ) -> str:
        client = Groq(api_key=self.api_key or GROQ_API_KEY)
        response = client.chat.completions.create(
            model=self.model,
            messages=[{'role': 'user', 'content': prompt}],
            temperature=0.0,
        )
        return response.choices[0].message.content

agent_llm = GroqLLM(model='llama-3.3-70b-versatile')

# Quick test
reply = agent_llm._call('Reply with only the word READY.')
print(f'LLM status: {reply.strip()}')
````

````python
# Part C: Step 2: ContractExtractor with Grounding Verification and Merge

import json
from dataclasses import dataclass

@dataclass
class ExtractionResult:
    fields:            ContractFields
    grounded:          dict   # field_name -> True/False
    evidence:          dict   # field_name -> text span that supports it
    unverified_fields: list   # fields where grounding failed

@dataclass
class MergeResult:
    fields:    ContractFields
    merge_log: list   # one entry per changed field


class ContractExtractor:

    # Keywords that ground each enum field value
    ENUM_KEYWORDS = {
        'ip_ownership': {
            'vendor': [
                'shall remain the property of vendor',
                'exclusive property of vendor',
                'vendor retains',
                'remain the exclusive property of vendor',
                'property of vendor',
            ],
            'client': [
                'exclusive property of client',
                'property of client',
                'owned by client',
                'shall be the exclusive property of client',
            ],
        },
        'termination_for_convenience': {
            'vendor_only': [
                'vendor may terminate',
                'vendor may terminate this agreement for any reason',
            ],
            'both': [
                'either party may terminate for convenience',
                'both parties may terminate',
            ],
            'none': [
                'neither party may terminate for convenience',
                'neither party may terminate',
            ],
        },
        'indemnification_direction': {
            'mutual':      ['each party shall indemnify', 'mutual indemnif'],
            'client_only': ['client shall indemnify', 'indemnify vendor'],
            'vendor_only': ['vendor shall indemnify', 'indemnify client'],
        },
        'dispute_resolution': {
            'arbitration': ['arbitration', 'arbitrator', 'jams'],
            'litigation':  ['litigation', 'court of law', 'courts of'],
        },
    }

    # Keywords that ground boolean fields
    BOOL_KEYWORDS = {
        'subcontractor_approval_required': ['subcontract'],
        'liability_cap_present':           ['liability', 'shall not exceed', 'limitation'],
        'auto_renewal':                    ['auto-renew', 'automatic renewal', 'auto renew'],
        'ip_covers_trained_models':        ['ml model', 'trained on', 'machine learning model'],
        'has_unresolved_conflict':         ['conflict', 'dispute', 'ambiguous'],
    }

    def __init__(self, llm: GroqLLM):
        self.llm = llm

    def extract(self, doc_text: str, doc_type: str) -> ExtractionResult:
        prompt = self._build_prompt(doc_text, doc_type)
        raw    = self.llm._call(prompt)

        # Strip markdown fences if present
        clean = raw.strip()
        if clean.startswith('```'):
            clean = '\n'.join(clean.split('\n')[1:])
        if clean.endswith('```'):
            clean = '\n'.join(clean.split('\n')[:-1])

        try:
            data = json.loads(clean.strip())
        except json.JSONDecodeError:
            data = {}

        fields   = self._dict_to_fields(data)
        grounded, evidence = self._verify_grounding(fields, doc_text)

        unverified = [
            fname for fname, ok in grounded.items()
            if not ok and getattr(fields, fname, None) is not None
        ]

        return ExtractionResult(
            fields            = fields,
            grounded          = grounded,
            evidence          = evidence,
            unverified_fields = unverified,
        )

    def merge_fields(self, base: ContractFields,
                     delta: ContractFields,
                     controlling_doc_id: str) -> MergeResult:
        import copy
        merged  = copy.deepcopy(base)
        log     = []

        for fname in base.__dataclass_fields__:
            delta_val = getattr(delta, fname, None)
            base_val  = getattr(base,  fname, None)
            if delta_val is not None and delta_val != base_val:
                setattr(merged, fname, delta_val)
                log.append(
                    f"{fname}: {base_val} -> {delta_val} "
                    f"(controlled by {controlling_doc_id})"
                )

        return MergeResult(fields=merged, merge_log=log)

    # ------------------------------------------------------------------
    # PROMPT BUILDER
    # ------------------------------------------------------------------

    def _build_prompt(self, doc_text: str, doc_type: str) -> str:
        amendment_instruction = ''
        if doc_type == 'Amendment':
            amendment_instruction = (
                '\nIMPORTANT: This is an Amendment. '
                'Extract ONLY the fields this Amendment explicitly modifies. '
                'Return null for ANY field not directly addressed by this document. '
                'Do not infer or fill in values from common contract patterns.\n'
            )

        return f"""You are a legal contract field extractor.
Extract the following fields from the contract text below.
Return ONLY valid JSON. No preamble, no markdown, no explanation.
Use null for any field not found in the document.
{amendment_instruction}
Fields to extract:
{{
  "governing_law":                  string or null,  // e.g. "California"
  "payment_terms_days":             integer or null, // e.g. 60
  "liability_cap_present":          boolean or null,
  "liability_cap_basis":            string or null,  // "fees_paid" or "fixed"
  "ip_ownership":                   string or null,  // "client" or "vendor"
  "auto_renewal":                   boolean or null,
  "auto_renewal_notice_days":       integer or null,
  "termination_for_convenience":    string or null,  // "none", "vendor_only", or "both"
  "termination_notice_days":        integer or null,
  "termination_notice_vendor_days": integer or null,
  "termination_notice_client_days": integer or null,
  "indemnification_direction":      string or null,  // "mutual", "client_only", "vendor_only"
  "non_compete_years":              float or null,
  "non_compete_scope":              string or null,
  "subcontractor_approval_required": boolean or null,
  "confidentiality_survival_years": float or null,
  "dispute_resolution":             string or null,  // "arbitration" or "litigation"
  "ip_covers_trained_models":       boolean or null,
  "has_unresolved_conflict":        boolean or null
}}

CONTRACT TEXT:
{doc_text}
"""

    # ------------------------------------------------------------------
    # FIELD CONVERSION
    # ------------------------------------------------------------------

    def _dict_to_fields(self, data: dict) -> ContractFields:
        def to_int(v):
            try: return int(v)
            except: return None
        def to_float(v):
            try: return float(v)
            except: return None
        def to_bool(v):
            if isinstance(v, bool): return v
            if isinstance(v, str):  return v.lower() == 'true'
            return None

        return ContractFields(
            governing_law                  = data.get('governing_law'),
            payment_terms_days             = to_int(data.get('payment_terms_days')),
            liability_cap_present          = to_bool(data.get('liability_cap_present')),
            liability_cap_basis            = data.get('liability_cap_basis'),
            ip_ownership                   = data.get('ip_ownership'),
            auto_renewal                   = to_bool(data.get('auto_renewal')),
            auto_renewal_notice_days       = to_int(data.get('auto_renewal_notice_days')),
            termination_for_convenience    = data.get('termination_for_convenience'),
            termination_notice_days        = to_int(data.get('termination_notice_days')),
            termination_notice_vendor_days = to_int(data.get('termination_notice_vendor_days')),
            termination_notice_client_days = to_int(data.get('termination_notice_client_days')),
            indemnification_direction      = data.get('indemnification_direction'),
            non_compete_years              = to_float(data.get('non_compete_years')),
            non_compete_scope              = data.get('non_compete_scope'),
            subcontractor_approval_required= to_bool(data.get('subcontractor_approval_required')),
            confidentiality_survival_years = to_float(data.get('confidentiality_survival_years')),
            dispute_resolution             = data.get('dispute_resolution'),
            ip_covers_trained_models       = to_bool(data.get('ip_covers_trained_models')) or False,
            has_unresolved_conflict        = to_bool(data.get('has_unresolved_conflict')) or False,
        )

    # ------------------------------------------------------------------
    # GROUNDING VERIFICATION
    # ------------------------------------------------------------------

    def _verify_grounding(self, fields: ContractFields,
                           doc_text: str) -> tuple:
        text_lower = doc_text.lower()
        grounded   = {}
        evidence   = {}

        for fname in fields.__dataclass_fields__:
            val = getattr(fields, fname, None)

            # G4: None values pass automatically
            if val is None:
                grounded[fname] = True
                evidence[fname] = ''
                continue

            # G1: Numeric fields — digit must appear in text
            if isinstance(val, (int, float)) and not isinstance(val, bool):
                str_val = str(int(val))
                if str_val in doc_text:
                    grounded[fname] = True
                    evidence[fname] = str_val
                else:
                    grounded[fname] = False
                    evidence[fname] = ''
                continue

            # G2: Enum fields — keyword map check
            if fname in self.ENUM_KEYWORDS and isinstance(val, str):
                keywords = self.ENUM_KEYWORDS[fname].get(val.lower(), [])
                found = next((kw for kw in keywords if kw in text_lower), None)
                grounded[fname] = found is not None
                evidence[fname] = found or ''
                continue

            # G3: Boolean fields — presence keyword check
            if fname in self.BOOL_KEYWORDS:
                keywords = self.BOOL_KEYWORDS[fname]
                found = next((kw for kw in keywords if kw in text_lower), None)
                grounded[fname] = found is not None
                evidence[fname] = found or ''
                continue

            # Default: string fields not in enum map — check if value appears in text
            if isinstance(val, str):
                grounded[fname] = val.lower() in text_lower
                evidence[fname] = val if val.lower() in text_lower else ''
                continue

            # Boolean fields not in BOOL_KEYWORDS — pass by default
            grounded[fname] = True
            evidence[fname] = ''

        return grounded, evidence
````

## What this part is about

In Lab 8a, you learned that an MCP server exposes tools the agent can discover and call without knowing how they are implemented. This part applies that pattern to a system where the tool selection decision actually matters. This is exactly how these systems work in production: the symbolic rule engine is typically owned and maintained by a separate team (often a legal engineering team), exposed as a service, and the AI application layer talks to it through a defined interface.

## The two-server architecture

Your server (`hw4_server.py`) handles everything built in this homework. The pre-built `mcp-server-fetch` from Lab 8a handles live HTTP calls. The `search_market_standard` tool in your server constructs a query and calls EDGAR or HuggingFace, but the actual HTTP request goes through `mcp-server-fetch`. The docstring must give the agent enough information to make a calibrated judgment. ````python
# Part D: Step 1: Write hw4_server.py

with open('hw4_server.py', 'w') as f:
    f.write(open('/tmp/hw4_server_source.py', 'w').name)

# Write the server directly as a string
server_code = f'''
import sys, json, asyncio, re, httpx
from dataclasses import dataclass, field
from typing import Optional
from mcp.server import Server
from mcp.server.stdio import stdio_server
from mcp import types
from groq import Groq
from langchain_core.language_models.llms import LLM
from langchain_core.callbacks.manager import CallbackManagerForLLMRun
from typing import Any, List

GROQ_API_KEY = "{GROQ_API_KEY}"

''' + '''
# ============================================================
# PART A: DocumentMeta, DocumentClassifier, ConflictReport, ConflictDetector
# ============================================================

@dataclass
class DocumentMeta:
    doc_id: str
    doc_type: str
    effective_date: Optional[str]
    parties: list
    precedence_claims: list
    clauses_covered: list

class DocumentClassifier:
    TYPE_KEYWORDS = {
        "Amendment": ["amendment","amended section","supersedes","replaces","deleted in its entirety","is hereby removed","effective immediately"],
        "Schedule":  ["schedule","exhibit","addendum","scope of services","services in scope","phase 1","phase 2","phase 3"],
        "MSA":       ["master services agreement","msa","agrees to provide","limitation of liability","indemnification","governing law"],
    }
    CLAUSE_KEYWORDS = {
        "payment_terms":      ["payment","invoice","net-30","days of receipt","late payment"],
        "ip_ownership":       ["intellectual property","work product","deliverable","ownership","exclusive property","ml model","trained on"],
        "governing_law":      ["governing law","governed by","laws of the state","jurisdiction"],
        "confidentiality":    ["confidential","confidentiality","hold in confidence"],
        "termination":        ["termination","terminate","non-renewal","notice of non-renewal"],
        "indemnification":    ["indemnif"],
        "liability_cap":      ["limitation of liability","liability shall not exceed","total liability","liability cap"],
        "dispute_resolution": ["arbitration","dispute","jams","mediation"],
        "non_compete":        ["non-compete","noncompete","compete","competitive"],
        "subcontractor":      ["subcontractor","subcontract","third-party provider"],
    }
    PRECEDENCE_PHRASES = ["shall control","notwithstanding","supersedes","replaces","deleted in its entirety","is hereby removed","takes precedence","governs over"]

    def classify(self, doc_id, text):
        text_lower = text.lower()
        return DocumentMeta(
            doc_id=doc_id,
            doc_type=self._detect_type(text_lower),
            effective_date=self._extract_date(text),
            parties=self._extract_parties(text),
            precedence_claims=self._extract_precedence_claims(text),
            clauses_covered=self._detect_clauses(text_lower),
        )

    def _detect_type(self, text_lower):
        first_lines = "\\n".join(text_lower.strip().splitlines()[:3])
        if "amendment" in first_lines: return "Amendment"
        if "schedule"  in first_lines: return "Schedule"
        if "master services agreement" in first_lines: return "MSA"
        for doc_type in ["Amendment","Schedule","MSA"]:
            if any(kw in text_lower for kw in self.TYPE_KEYWORDS[doc_type]):
                return doc_type
        return "Unknown"

    def _extract_date(self, text):
        m = re.search(r"(?:as of|entered into|effective)\\s+([A-Z][a-z]+ \\d{1,2},\\s*\\d{4})", text)
        return m.group(1) if m else None

    def _extract_parties(self, text):
        parties = re.findall(r\'\"([^"]+?)\"\\s*\\(\\s*"(?:Client|Vendor|Company|Contractor|Party)"\\s*\\)\', text)
        return list(dict.fromkeys(parties))

    def _extract_precedence_claims(self, text):
        return [s.strip() for s in re.split(r"(?<=[.!?])\\s+", text)
                if any(p in s.lower() for p in self.PRECEDENCE_PHRASES)]

    def _detect_clauses(self, text_lower):
        return [c for c, kws in self.CLAUSE_KEYWORDS.items() if any(kw in text_lower for kw in kws)]


@dataclass
class ConflictReport:
    conflicts: list
    has_conflicts: bool
    resolution_confidence: str

class ConflictDetector:
    FIELD_TO_CLAUSE = {
        "payment_terms_days":          "payment_terms",
        "ip_ownership":                "ip_ownership",
        "ip_covers_trained_models":    "ip_ownership",
        "termination_for_convenience": "termination",
        "termination_notice_days":     "termination",
        "governing_law":               "governing_law",
        "liability_cap_present":       "liability_cap",
        "non_compete_years":           "non_compete",
    }
    CLAUSE_KEYWORDS = {
        "payment_terms": ["payment","invoice"],
        "ip_ownership":  ["ip","intellectual property","model","deliverable"],
        "termination":   ["terminat"],
        "governing_law": ["governing","jurisdiction"],
    }

    def detect(self, meta_a, fields_a, meta_b, fields_b):
        conflicts = []
        for fname, clause in self.FIELD_TO_CLAUSE.items():
            val_a = fields_a.get(fname)
            val_b = fields_b.get(fname)
            if val_a is None or val_b is None or val_a == val_b:
                continue
            ctrl_doc, ctrl_val, rationale, confidence = self._resolve(fname, clause, meta_a, val_a, meta_b, val_b)
            conflicts.append({
                "clause": clause, "field": fname,
                "doc_a": meta_a.doc_id, "doc_a_value": val_a,
                "doc_b": meta_b.doc_id, "doc_b_value": val_b,
                "controlling_doc": ctrl_doc, "controlling_value": ctrl_val,
                "rationale": rationale, "confidence": confidence,
                "risk_amplified": self._is_risk_amplified(fname, ctrl_val, val_a, val_b),
            })
        confidence = self._overall_confidence(conflicts)
        return ConflictReport(conflicts=conflicts, has_conflicts=len(conflicts)>0, resolution_confidence=confidence)

    def _claims_control(self, meta, clause):
        kws = self.CLAUSE_KEYWORDS.get(clause, [])
        return any(any(kw in c.lower() for kw in kws) for c in meta.precedence_claims)

    def _resolve(self, fname, clause, meta_a, val_a, meta_b, val_b):
        a_claims = self._claims_control(meta_a, clause)
        b_claims = self._claims_control(meta_b, clause)
        if a_claims and b_claims:
            return (meta_b.doc_id, val_b,
                f"AMBIGUOUS: Both {meta_a.doc_id} and {meta_b.doc_id} claim precedence over {clause}. Legal review required.",
                "LOW")
        if meta_a.doc_type == "Amendment" and clause in meta_a.clauses_covered:
            return (meta_a.doc_id, val_a, f"Amendment {meta_a.doc_id} explicitly replaces {clause}.", "HIGH")
        if meta_b.doc_type == "Amendment" and clause in meta_b.clauses_covered:
            return (meta_b.doc_id, val_b, f"Amendment {meta_b.doc_id} explicitly replaces {clause}.", "HIGH")
        if meta_a.doc_type == "Schedule" and a_claims:
            return (meta_a.doc_id, val_a, f"Schedule {meta_a.doc_id} notwithstanding claim over {clause}.", "MEDIUM")
        if meta_b.doc_type == "Schedule" and b_claims:
            return (meta_b.doc_id, val_b, f"Schedule {meta_b.doc_id} notwithstanding claim over {clause}.", "MEDIUM")
        later = meta_b if meta_b.doc_id > meta_a.doc_id else meta_a
        later_val = val_b if later == meta_b else val_a
        return (later.doc_id, later_val, f"Later document ({later.doc_id}) controls by default.", "MEDIUM")

    def _is_risk_amplified(self, fname, ctrl_val, val_a, val_b):
        if fname == "payment_terms_days":
            other = val_a if ctrl_val == val_b else val_b
            return isinstance(ctrl_val,(int,float)) and isinstance(other,(int,float)) and ctrl_val > other
        if fname in ("ip_ownership","ip_covers_trained_models"):
            return str(ctrl_val).lower() in ("vendor","true")
        if fname == "termination_for_convenience":
            return str(ctrl_val).lower() == "vendor_only"
        return False

    def _overall_confidence(self, conflicts):
        if not conflicts: return "HIGH"
        confs = [c["confidence"] for c in conflicts]
        if "LOW" in confs: return "LOW"
        if "MEDIUM" in confs: return "MEDIUM"
        return "HIGH"


# ============================================================
# PART B: ContractFields, JurisdictionRuleSet, Factory, Engine
# ============================================================

@dataclass
class ContractFields:
    governing_law: Optional[str] = None
    payment_terms_days: Optional[int] = None
    liability_cap_present: Optional[bool] = None
    liability_cap_basis: Optional[str] = None
    ip_ownership: Optional[str] = None
    auto_renewal: Optional[bool] = None
    auto_renewal_notice_days: Optional[int] = None
    termination_for_convenience: Optional[str] = None
    termination_notice_days: Optional[int] = None
    indemnification_direction: Optional[str] = None
    non_compete_years: Optional[float] = None
    non_compete_scope: Optional[str] = None
    subcontractor_approval_required: Optional[bool] = None
    confidentiality_survival_years: Optional[float] = None
    dispute_resolution: Optional[str] = None
    controlling_payment_terms_days: Optional[int] = None
    controlling_document_id: Optional[str] = None
    has_unresolved_conflict: bool = False
    ip_covers_trained_models: bool = False
    termination_notice_vendor_days: Optional[int] = None
    termination_notice_client_days: Optional[int] = None

@dataclass
class JurisdictionRuleSet:
    jurisdiction: str
    payment_medium_threshold_days: int
    payment_high_threshold_days: int
    non_compete_base_level: str
    trained_model_ip_extra_flag: Optional[str]

class JurisdictionRuleSetFactory:
    @classmethod
    def get(cls, governing_law):
        if governing_law and "california" in governing_law.lower():
            return JurisdictionRuleSet("California", 45, 75, "HIGH",
                "CCPA risk: Vendor use of Client data in commercial models may require Client consent under Cal. Civ. Code §1798.100")
        if governing_law and "texas" in governing_law.lower():
            return JurisdictionRuleSet("Texas", 60, 90, "MEDIUM", None)
        return JurisdictionRuleSet(governing_law or "Unknown", 60, 90, "MEDIUM", None)

class ContractRuleEngine:
    RULE_VERSION = "1.0.0"
    def __init__(self, ruleset):
        self.ruleset = ruleset
    def generate_report(self, fields):
        tier1 = [r for r in [self._r01(fields),self._r02(fields),self._r03(fields),
            self._r04(fields),self._r05(fields),self._r06(fields),self._r07(fields),
            self._r08(fields),self._r09(fields),self._r10(fields),self._r11(fields),
            self._r12(fields)] if r]
        tier2 = [r for r in [self._c01(fields,{x["field"] for x in tier1}),
            self._c02(fields,{x["field"] for x in tier1}),
            self._c03(fields,{x["field"] for x in tier1})] if r]
        levels = [f["level"] for f in tier1+tier2]
        overall = "CRITICAL" if "CRITICAL" in levels else "HIGH" if "HIGH" in levels else "MEDIUM" if "MEDIUM" in levels else "LOW"
        return {"overall_risk":overall,"tier1_findings":tier1,"tier2_findings":tier2,
                "jurisdiction":self.ruleset.jurisdiction,"critical_count":levels.count("CRITICAL"),
                "high_count":levels.count("HIGH"),"medium_count":levels.count("MEDIUM"),"rule_version":self.RULE_VERSION}
    def _r01(self,f):
        days = f.controlling_payment_terms_days or f.payment_terms_days
        if days is None: return None
        if days > self.ruleset.payment_high_threshold_days: level="HIGH"
        elif days > self.ruleset.payment_medium_threshold_days: level="MEDIUM"
        else: return None
        return {"field":"R01","level":level,"explanation":f"Payment terms of {days} days exceed the {self.ruleset.jurisdiction} {level.lower()} threshold of {self.ruleset.payment_high_threshold_days if level=='HIGH' else self.ruleset.payment_medium_threshold_days} days."}
    def _r02(self,f):
        if f.liability_cap_present is False: return {"field":"R02","level":"HIGH","explanation":"No liability cap present. Client exposure to damages is unlimited."}
        if f.liability_cap_present and f.liability_cap_basis not in ("fees_paid","fixed"): return {"field":"R02","level":"MEDIUM","explanation":f"Liability cap basis is vague: {f.liability_cap_basis}."}
        return None
    def _r03(self,f):
        if f.ip_ownership=="vendor": return {"field":"R03","level":"HIGH","explanation":"IP ownership of custom deliverables assigned to Vendor. Client pays for work it does not own."}
        return None
    def _r04(self,f):
        if not f.ip_covers_trained_models: return None
        exp = "ML model IP assigned to Vendor. Models may encode proprietary Client insights."
        if self.ruleset.trained_model_ip_extra_flag: exp += " " + self.ruleset.trained_model_ip_extra_flag
        return {"field":"R04","level":"HIGH","explanation":exp}
    def _r05(self,f):
        if f.auto_renewal and f.auto_renewal_notice_days and f.auto_renewal_notice_days < 14:
            return {"field":"R05","level":"MEDIUM","explanation":f"Auto-renewal notice of {f.auto_renewal_notice_days} days is below the 14-day minimum."}
        return None
    def _r06(self,f):
        if f.termination_for_convenience=="vendor_only": return {"field":"R06","level":"HIGH","explanation":"Vendor may terminate for any reason. Client has no equivalent right."}
        return None
    def _r07(self,f):
        v,c = f.termination_notice_vendor_days, f.termination_notice_client_days
        if v and c and v<14 and c>30: return {"field":"R07","level":"HIGH","explanation":f"Termination asymmetry: Vendor exits in {v} days, Client requires {c} days to terminate."}
        return None
    def _r08(self,f):
        if f.indemnification_direction=="client_only": return {"field":"R08","level":"HIGH","explanation":"Indemnification runs only against Client. Vendor bears no obligations."}
        return None
    def _r09(self,f):
        if f.non_compete_years is None: return None
        basis = "Cal. B&P Code §16600 renders most non-competes void." if self.ruleset.jurisdiction=="California" else "Enforceable if reasonable."
        return {"field":"R09","level":self.ruleset.non_compete_base_level,"explanation":f"Non-compete of {f.non_compete_years}yr ({f.non_compete_scope}) present. {basis}"}
    def _r10(self,f):
        if f.subcontractor_approval_required is False: return {"field":"R10","level":"MEDIUM","explanation":"Vendor may engage subcontractors without Client approval."}
        return None
    def _r11(self,f):
        if f.governing_law is None: return {"field":"R11","level":"MEDIUM","explanation":"No governing law specified across all documents."}
        return None
    def _r12(self,f):
        if f.has_unresolved_conflict: return {"field":"R12","level":"MEDIUM","explanation":"Unresolved cross-document conflict. Legal ambiguity about which value controls."}
        return None
    def _c01(self,f,fired):
        if "R06" in fired and "R08" in fired and ("R03" in fired or "R04" in fired):
            return {"field":"C01","level":"CRITICAL","explanation":"STRUCTURAL IMBALANCE: Vendor controls termination, owns IP, bears no indemnity. Client has all obligations and no protections."}
        return None
    def _c02(self,f,fired):
        if "R04" in fired and "R10" in fired:
            return {"field":"C02","level":"CRITICAL","explanation":"DATA SOVEREIGNTY RISK: Vendor owns ML models trained on Client data and can share pipeline with unapproved subcontractors."}
        return None
    def _c03(self,f,fired):
        if "R01" in fired and "R06" in fired:
            return {"field":"C03","level":"CRITICAL","explanation":"PAYMENT AND TERMINATION TRAP: Client owes extended payment after milestones but Vendor can exit in 7 days."}
        return None


# ============================================================
# PART C: GroqLLM, ExtractionResult, MergeResult, ContractExtractor
# ============================================================

class GroqLLM(LLM):
    model: str = "llama-3.3-70b-versatile"
    api_key: str = ""
    def __init__(self, model, api_key="", **kwargs):
        super().__init__(model=model, api_key=api_key, **kwargs)
    @property
    def _llm_type(self): return "groq"
    def _call(self, prompt, stop=None, run_manager=None, **kwargs):
        client = Groq(api_key=self.api_key or GROQ_API_KEY)
        response = client.chat.completions.create(
            model=self.model,
            messages=[{"role":"user","content":prompt}],
            temperature=0.0,
        )
        return response.choices[0].message.content

@dataclass
class ExtractionResult:
    fields: ContractFields
    grounded: dict
    evidence: dict
    unverified_fields: list

@dataclass
class MergeResult:
    fields: ContractFields
    merge_log: list

class ContractExtractor:
    ENUM_KEYWORDS = {
        "ip_ownership": {
            "vendor": ["shall remain the property of vendor","exclusive property of vendor","vendor retains","remain the exclusive property of vendor","property of vendor"],
            "client": ["exclusive property of client","property of client","owned by client","shall be the exclusive property of client"],
        },
        "termination_for_convenience": {
            "vendor_only": ["vendor may terminate","vendor may terminate this agreement for any reason"],
            "both":        ["either party may terminate for convenience","both parties may terminate"],
            "none":        ["neither party may terminate for convenience","neither party may terminate"],
        },
        "indemnification_direction": {
            "mutual":      ["each party shall indemnify","mutual indemnif"],
            "client_only": ["client shall indemnify","indemnify vendor"],
            "vendor_only": ["vendor shall indemnify","indemnify client"],
        },
        "dispute_resolution": {
            "arbitration": ["arbitration","arbitrator","jams"],
            "litigation":  ["litigation","court of law","courts of"],
        },
    }
    BOOL_KEYWORDS = {
        "subcontractor_approval_required": ["subcontract"],
        "liability_cap_present":           ["liability","shall not exceed","limitation"],
        "auto_renewal":                    ["auto-renew","automatic renewal","auto renew"],
        "ip_covers_trained_models":        ["ml model","trained on","machine learning model"],
        "has_unresolved_conflict":         ["conflict","dispute","ambiguous"],
    }

    def __init__(self, llm):
        self.llm = llm

    def extract(self, doc_text, doc_type):
        prompt = self._build_prompt(doc_text, doc_type)
        raw = self.llm._call(prompt)
        clean = raw.strip()
        if clean.startswith("```"): clean = "\\n".join(clean.split("\\n")[1:])
        if clean.endswith("```"):   clean = "\\n".join(clean.split("\\n")[:-1])
        try:    data = json.loads(clean.strip())
        except: data = {}
        fields = self._dict_to_fields(data)
        grounded, evidence = self._verify_grounding(fields, doc_text)
        unverified = [fn for fn,ok in grounded.items() if not ok and getattr(fields,fn,None) is not None]
        return ExtractionResult(fields=fields, grounded=grounded, evidence=evidence, unverified_fields=unverified)

    def merge_fields(self, base, delta, controlling_doc_id):
        import copy
        merged = copy.deepcopy(base)
        log = []
        for fname in base.__dataclass_fields__:
            dv = getattr(delta, fname, None)
            bv = getattr(base,  fname, None)
            if dv is not None and dv != bv:
                setattr(merged, fname, dv)
                log.append(f"{fname}: {bv} -> {dv} (controlled by {controlling_doc_id})")
        return MergeResult(fields=merged, merge_log=log)

    def _build_prompt(self, doc_text, doc_type):
        amendment_instruction = ""
        if doc_type == "Amendment":
            amendment_instruction = "\\nIMPORTANT: This is an Amendment. Extract ONLY the fields this Amendment explicitly modifies. Return null for ANY field not directly addressed. Do not infer values from common patterns.\\n"
        return f"""You are a legal contract field extractor.
Extract the following fields from the contract text below.
Return ONLY valid JSON. No preamble, no markdown, no explanation.
Use null for any field not found in the document.
{amendment_instruction}
Fields:
{{
  "governing_law": string or null,
  "payment_terms_days": integer or null,
  "liability_cap_present": boolean or null,
  "liability_cap_basis": string or null,
  "ip_ownership": string or null,
  "auto_renewal": boolean or null,
  "auto_renewal_notice_days": integer or null,
  "termination_for_convenience": string or null,
  "termination_notice_days": integer or null,
  "termination_notice_vendor_days": integer or null,
  "termination_notice_client_days": integer or null,
  "indemnification_direction": string or null,
  "non_compete_years": float or null,
  "non_compete_scope": string or null,
  "subcontractor_approval_required": boolean or null,
  "confidentiality_survival_years": float or null,
  "dispute_resolution": string or null,
  "ip_covers_trained_models": boolean or null,
  "has_unresolved_conflict": boolean or null
}}

CONTRACT TEXT:
{doc_text}
"""

    def _dict_to_fields(self, data):
        def ti(v):
            try: return int(v)
            except: return None
        def tf(v):
            try: return float(v)
            except: return None
        def tb(v):
            if isinstance(v,bool): return v
            if isinstance(v,str):  return v.lower()=="true"
            return None
        return ContractFields(
            governing_law=data.get("governing_law"),
            payment_terms_days=ti(data.get("payment_terms_days")),
            liability_cap_present=tb(data.get("liability_cap_present")),
            liability_cap_basis=data.get("liability_cap_basis"),
            ip_ownership=data.get("ip_ownership"),
            auto_renewal=tb(data.get("auto_renewal")),
            auto_renewal_notice_days=ti(data.get("auto_renewal_notice_days")),
            termination_for_convenience=data.get("termination_for_convenience"),
            termination_notice_days=ti(data.get("termination_notice_days")),
            termination_notice_vendor_days=ti(data.get("termination_notice_vendor_days")),
            termination_notice_client_days=ti(data.get("termination_notice_client_days")),
            indemnification_direction=data.get("indemnification_direction"),
            non_compete_years=tf(data.get("non_compete_years")),
            non_compete_scope=data.get("non_compete_scope"),
            subcontractor_approval_required=tb(data.get("subcontractor_approval_required")),
            confidentiality_survival_years=tf(data.get("confidentiality_survival_years")),
            dispute_resolution=data.get("dispute_resolution"),
            ip_covers_trained_models=tb(data.get("ip_covers_trained_models")) or False,
            has_unresolved_conflict=tb(data.get("has_unresolved_conflict")) or False,
        )

    def _verify_grounding(self, fields, doc_text):
        text_lower = doc_text.lower()
        grounded, evidence = {}, {}
        for fname in fields.__dataclass_fields__:
            val = getattr(fields, fname, None)
            if val is None:
                grounded[fname]=True; evidence[fname]=""; continue
            if isinstance(val,(int,float)) and not isinstance(val,bool):
                sv = str(int(val))
                grounded[fname] = sv in doc_text; evidence[fname] = sv if sv in doc_text else ""; continue
            if fname in self.ENUM_KEYWORDS and isinstance(val,str):
                kws = self.ENUM_KEYWORDS[fname].get(val.lower(),[])
                found = next((k for k in kws if k in text_lower), None)
                grounded[fname]=found is not None; evidence[fname]=found or ""; continue
            if fname in self.BOOL_KEYWORDS:
                kws = self.BOOL_KEYWORDS[fname]
                found = next((k for k in kws if k in text_lower), None)
                grounded[fname]=found is not None; evidence[fname]=found or ""; continue
            if isinstance(val,str):
                grounded[fname]=val.lower() in text_lower; evidence[fname]=val if val.lower() in text_lower else ""; continue
            grounded[fname]=True; evidence[fname]=""
        return grounded, evidence


# ============================================================
# MCP SERVER
# ============================================================

app = Server("hw4_server")

@app.list_tools()
async def list_tools():
    return [
        types.Tool(
            name="identify_document",
            description="Identify the type of a legal document (MSA, Amendment, or Schedule) using keyword and structural heuristics. Returns document metadata including doc_type, effective_date, clauses_covered, and any precedence claims. Always call this first for every document before any other tool. The doc_type determines how extraction and conflict detection proceed.",
            inputSchema={"type":"object","properties":{"doc_id":{"type":"string"},"doc_text":{"type":"string"}},"required":["doc_id","doc_text"]},
        ),
        types.Tool(
            name="extract_fields",
            description="Extract structured contract fields from raw document text using an LLM. Amendments are extracted as deltas — only fields explicitly modified are returned as non-null. Call after identify_document and pass the doc_type. Returns fields, grounded status per field, and unverified fields where grounding failed and values should not be trusted.",
            inputSchema={"type":"object","properties":{"doc_id":{"type":"string"},"doc_text":{"type":"string"},"doc_type":{"type":"string"}},"required":["doc_id","doc_text","doc_type"]},
        ),
        types.Tool(
            name="detect_conflict",
            description="Compare extracted fields from two documents and identify conflicts where both documents address the same clause with different values. Call only when two documents cover the same clause type — do not call for every possible pair. Returns conflicts list, has_conflicts flag, and resolution_confidence. LOW confidence means genuinely ambiguous and requires legal review.",
            inputSchema={"type":"object","properties":{"doc_id_a":{"type":"string"},"doc_type_a":{"type":"string"},"fields_json_a":{"type":"string"},"precedence_claims_a":{"type":"array","items":{"type":"string"}},"clauses_covered_a":{"type":"array","items":{"type":"string"}},"doc_id_b":{"type":"string"},"doc_type_b":{"type":"string"},"fields_json_b":{"type":"string"},"precedence_claims_b":{"type":"array","items":{"type":"string"}},"clauses_covered_b":{"type":"array","items":{"type":"string"}}},"required":["doc_id_a","doc_type_a","fields_json_a","precedence_claims_a","clauses_covered_a","doc_id_b","doc_type_b","fields_json_b","precedence_claims_b","clauses_covered_b"]},
        ),
        types.Tool(
            name="resolve_conflict",
            description="Apply precedence logic to a ConflictReport and return the controlling field values after resolution. Call after detect_conflict when has_conflicts is True. Returns resolved_fields with winning values and rationale_log explaining which document controls each clause. LOW confidence resolutions must be flagged in the final brief as legally ambiguous.",
            inputSchema={"type":"object","properties":{"conflict_json":{"type":"string"}},"required":["conflict_json"]},
        ),
        types.Tool(
            name="check_rules",
            description="Run the two-tier deterministic rule engine on resolved contract fields. Returns tier1_findings, tier2_findings that fire CRITICAL when multiple HIGH risks combine, overall_risk level, and counts. Do NOT call before governing law is identified from all documents — jurisdiction determines which thresholds apply. Pass governing_law as unknown only if no document specifies it.",
            inputSchema={"type":"object","properties":{"fields_json":{"type":"string"},"governing_law":{"type":"string"}},"required":["fields_json","governing_law"]},
        ),
        types.Tool(
            name="get_rule_explanation",
            description="Return the full explanation and legal basis for a specific rule ID and jurisdiction. Call for every CRITICAL and HIGH finding when producing a client-facing brief. Explanation includes jurisdiction-specific threshold and the statute behind it such as Cal. B&P Code 16600 for non-competes in California. Rule IDs are R01-R12 for tier1 and C01-C03 for tier2.",
            inputSchema={"type":"object","properties":{"rule_id":{"type":"string"},"jurisdiction":{"type":"string"}},"required":["rule_id","jurisdiction"]},
        ),
        types.Tool(
            name="search_market_standard",
            description="Query SEC EDGAR and HuggingFace CUAD dataset to estimate how common a clause type is in real contracts from the same jurisdiction. Call ONLY for MEDIUM findings or HIGH findings where risk is jurisdiction-dependent and market prevalence would change the recommendation. Do NOT call for clear statutory violations such as California non-compete which is void under Cal. B&P Code 16600, for LOW findings, or where threshold is not a matter of market norms. Returns prevalence_estimate, sample_count, market_assessment, and source.",
            inputSchema={"type":"object","properties":{"clause_type":{"type":"string"},"jurisdiction":{"type":"string"},"finding_level":{"type":"string"}},"required":["clause_type","jurisdiction","finding_level"]},
        ),
    ]

@app.call_tool()
async def call_tool(name: str, arguments: dict):
    try:
        result = await _dispatch(name, arguments)
        return [types.TextContent(type="text", text=json.dumps(result, indent=2))]
    except Exception as e:
        return [types.TextContent(type="text", text=json.dumps({"error": str(e)}))]

async def _dispatch(name, arguments):
    if name == "identify_document":
        clf = DocumentClassifier()
        meta = clf.classify(arguments["doc_id"], arguments["doc_text"])
        return {"doc_id":meta.doc_id,"doc_type":meta.doc_type,"effective_date":meta.effective_date,
                "parties":meta.parties,"clauses_covered":meta.clauses_covered,"precedence_claims":meta.precedence_claims}

    elif name == "extract_fields":
        llm = GroqLLM(model="llama-3.3-70b-versatile", api_key=GROQ_API_KEY)
        ext = ContractExtractor(llm=llm)
        res = ext.extract(arguments["doc_text"], arguments["doc_type"])
        return {"fields":{f:getattr(res.fields,f) for f in res.fields.__dataclass_fields__},
                "grounded":res.grounded,"unverified_fields":res.unverified_fields}

    elif name == "detect_conflict":
        meta_a = DocumentMeta(doc_id=arguments["doc_id_a"],doc_type=arguments["doc_type_a"],
            effective_date=None,parties=[],precedence_claims=arguments["precedence_claims_a"],
            clauses_covered=arguments["clauses_covered_a"])
        meta_b = DocumentMeta(doc_id=arguments["doc_id_b"],doc_type=arguments["doc_type_b"],
            effective_date=None,parties=[],precedence_claims=arguments["precedence_claims_b"],
            clauses_covered=arguments["clauses_covered_b"])
        fields_a = json.loads(arguments["fields_json_a"])
        fields_b = json.loads(arguments["fields_json_b"])
        report = ConflictDetector().detect(meta_a,fields_a,meta_b,fields_b)
        return {"conflicts":report.conflicts,"has_conflicts":report.has_conflicts,
                "resolution_confidence":report.resolution_confidence}

    elif name == "resolve_conflict":
        data = json.loads(arguments["conflict_json"])
        resolved, log = {}, []
        for c in data.get("conflicts",[]):
            resolved[c["field"]] = c["controlling_value"]
            log.append(f"{c[\'field\']} ({c[\'clause\']}): {c[\'controlling_doc\']} controls = {c[\'controlling_value\']} — {c[\'rationale\']}")
        return {"resolved_fields":resolved,"rationale_log":log}

    elif name == "check_rules":
        fd = json.loads(arguments["fields_json"])
        fields = ContractFields(**{k:v for k,v in fd.items() if k in ContractFields.__dataclass_fields__})
        engine = ContractRuleEngine(JurisdictionRuleSetFactory.get(arguments.get("governing_law")))
        return engine.generate_report(fields)

    elif name == "get_rule_explanation":
        rule_id = arguments["rule_id"]
        jur = arguments.get("jurisdiction","unknown")
        rs = JurisdictionRuleSetFactory.get(jur)
        explanations = {
            "R01":f"Payment terms exceed the {jur} threshold of {rs.payment_high_threshold_days} days. California Prompt Payment Act sets a 30-day statutory default.",
            "R02":"Liability cap absent or vague. Standard practice requires a cap tied to fees paid in the preceding 12 months.",
            "R03":"IP on custom deliverables assigned to Vendor. Client pays for work it does not own.",
            "R04":f"ML model IP to Vendor. {rs.trained_model_ip_extra_flag or ''}",
            "R05":"Auto-renewal notice under 14 days gives Client insufficient time to evaluate renewal.",
            "R06":"Vendor-only termination for convenience creates severe power imbalance.",
            "R07":"Termination asymmetry: Vendor exits in days, Client locked in for months.",
            "R08":"One-sided indemnification leaves Client exposed to all third-party claims.",
            "R09":f"Non-compete clause. {jur} base level: {rs.non_compete_base_level}. {'Cal. B&P Code §16600 renders most non-competes void.' if jur=='California' else ''}",
            "R10":"Subcontractor engagement without Client approval. No visibility into data handling.",
            "R11":"No governing law specified. Creates uncertainty about applicable jurisdiction.",
            "R12":"Unresolved cross-document conflict. Legal ambiguity about which value controls.",
            "C01":"CRITICAL: Structural imbalance — Vendor controls termination, owns IP, bears no indemnity.",
            "C02":"CRITICAL: Data sovereignty risk — Vendor owns ML models, can share with unapproved subcontractors.",
            "C03":"CRITICAL: Payment and termination trap — Client owes payment but Vendor can exit in 7 days.",
        }
        return {"rule_id":rule_id,"jurisdiction":jur,"explanation":explanations.get(rule_id,f"No explanation for {rule_id}")}

    elif name == "search_market_standard":
        clause_type = arguments["clause_type"]
        jurisdiction = arguments["jurisdiction"]
        finding_level = arguments["finding_level"]
        result = {"prevalence_estimate":"unknown","sample_count":0,"market_assessment":"API unavailable","source":"none"}
        try:
            async with httpx.AsyncClient(timeout=10) as client:
                r = await client.get(
                    "https://datasets-server.huggingface.co/search",
                    params={"dataset":"theatticusproject/cuad","config":"default","split":"train",
                            "query":clause_type,"offset":0,"length":5},
                    headers={"User-Agent":"ClearClauseAI student@university.edu"},
                )
                if r.status_code == 200:
                    rows = r.json().get("rows",[])
                    result = {"prevalence_estimate":f"{len(rows)} matching clauses found",
                              "sample_count":len(rows),
                              "market_assessment":f"{clause_type} appears in {len(rows)} sampled contracts. Market context for {jurisdiction} {finding_level} finding.",
                              "source":"HuggingFace CUAD dataset"}
        except Exception as e:
            result["market_assessment"] = f"Search unavailable: {str(e)}"
        return result

    return {"error": f"Unknown tool: {name}"}

async def main():
    async with stdio_server() as (read_stream, write_stream):
        await app.run(read_stream, write_stream, app.create_initialization_options())

if __name__ == "__main__":
    asyncio.run(main())
'''

with open('hw4_server.py', 'w') as f:
    f.write(server_code)

import py_compile, os
try:
    py_compile.compile('hw4_server.py', doraise=True)
    print(f'hw4_server.py: valid Python ({os.path.getsize("hw4_server.py")} bytes)')
except py_compile.PyCompileError as e:
    print(f'Syntax error: {e}')
````

# Part E: ReAct Agent (10 points)

## What this part is about

This is where all five parts come together. The agent receives the full document package and a goal: produce a risk brief the client's legal team can act on, and must figure out on its own how to accomplish that goal using the tools available.

There is no step-by-step pipeline in the task prompt. The agent is not told to call `identify_document` first, or to wait for governing law before calling `check_rules`. It must reason about the documents, discover their structure, track what it has learned, and decide what to do next based on what it finds. This is what distinguishes an agent from a scripted workflow.

## What makes the tool call sequence unpredictable

The governing law only appears in the Schedule. Until the agent reads the Schedule and calls `identify_document` on it, it does not know which jurisdiction rules to apply. An agent that calls `check_rules` before reading all three documents will use the wrong thresholds.

The conflict detection calls are conditional. The agent should call `detect_conflict` when it has extracted fields from two documents that both cover the same clause type. How many such pairs exist depends on what the documents contain, the agent must reason about this, not assume a fixed number of comparisons.

The market standard search is expensive in terms of iterations and real network calls. The agent must decide whether a given finding is the kind where market data would change the recommendation. For a California non-compete, market data is irrelevant, the clause is void by statute. For a 90-day payment term under California's Prompt Payment Act framework, market data may help contextualize whether this is unusually long for the software services sector. The agent needs to make these distinctions.

## The ablation test

After running the full three-document analysis, you will run the same agent on just the MSA and Amendment (no Schedule). Several things should change: governing law cannot be determined, one of the conflict comparisons no longer exists, and the IP override from the Schedule is absent. Your written analysis in Part F Question 1 asks you to document exactly what changed and why. If nothing changes, the agent is not doing real reasoning.

**Grading rubric:** Agent identifies all three conflicts in the three-document run (4 pts). Agent calls `search_market_standard` selectively, not for every finding (3 pts). Agent correctly applies California jurisdiction rules to at least one finding (3 pts).

````python
# Part E: Step 1: Infrastructure from Lab 8a (fixed for multi-param tools)

import sys, os, re, asyncio, json
import nest_asyncio
nest_asyncio.apply()

try:
    sys.stderr.fileno()
except Exception:
    sys.stderr = open(os.devnull, 'w')

from mcp import ClientSession, StdioServerParameters
from mcp.client.stdio import stdio_client
from langchain_core.tools import Tool
from langchain_core.prompts import PromptTemplate


class MCPToolBridge:
    def __init__(self, server_script: str):
        self.server_script = server_script
        self._tool_schemas = {}
        self.server_params = StdioServerParameters(
            command=sys.executable,
            args=[server_script],
            env=None
        )

    async def _call_tool_async(self, tool_name: str, tool_input: str) -> str:
        # Try to parse as JSON dict for multi-param tools
        try:
            parsed = json.loads(tool_input)
            if isinstance(parsed, dict):
                arguments = parsed
            else:
                param_name = self._tool_schemas.get(tool_name, 'input')
                arguments = {param_name: tool_input}
        except (json.JSONDecodeError, ValueError):
            param_name = self._tool_schemas.get(tool_name, 'input')
            arguments = {param_name: tool_input}

        async with stdio_client(self.server_params) as (read, write):
            async with ClientSession(read, write) as session:
                await session.initialize()
                result = await session.call_tool(tool_name, arguments)
                if result.isError:
                    return 'Tool error: ' + result.content[0].text
                return result.content[0].text

    def _call_tool_sync(self, tool_name: str, tool_input: str) -> str:
        loop = asyncio.get_event_loop()
        return loop.run_until_complete(
            self._call_tool_async(tool_name, tool_input)
        )

    async def get_langchain_tools(self) -> list:
        async with stdio_client(self.server_params) as (read, write):
            async with ClientSession(read, write) as session:
                await session.initialize()
                tools_result = await session.list_tools()
                langchain_tools = []
                for mcp_tool in tools_result.tools:
                    props = mcp_tool.inputSchema.get('properties', {})
                    first_param = next(iter(props), 'input')
                    self._tool_schemas[mcp_tool.name] = first_param
                    def make_func(name):
                        def tool_func(input_str: str) -> str:
                            return self._call_tool_sync(name, input_str)
                        return tool_func
                    langchain_tools.append(Tool(
                        name=mcp_tool.name,
                        description=mcp_tool.description,
                        func=make_func(mcp_tool.name)
                    ))
                return langchain_tools


class MCPModuleBridge(MCPToolBridge):
    def __init__(self, module_name: str):
        self.server_script = module_name
        self._tool_schemas = {}
        self.server_params = StdioServerParameters(
            command=sys.executable,
            args=['-m', module_name],
            env=None
        )


REACT_TEMPLATE = """You are a contract risk analyst. Answer the question using the tools available.

Tools available:
{tools}

Use this format exactly:
Thought: what you are thinking
Action: tool_name
Action Input: input to the tool as a JSON object
Observation: result of the tool
... (repeat Thought/Action/Action Input/Observation as needed)
Thought: I now know the final answer
Final Answer: your complete answer

Important rules:
- Always call identify_document for EVERY document before anything else
- Always pass inputs as JSON objects with the correct field names
- Always call extract_fields after identify_document for each document
- Do NOT call check_rules until you have read ALL documents and know the governing law
- Call detect_conflict only when two documents share the same clause type
- Call search_market_standard only for MEDIUM findings or borderline HIGH findings
- Never call search_market_standard for statutory violations (e.g. California non-compete)

Question: {input}

{agent_scratchpad}"""

react_prompt = PromptTemplate(
    input_variables=['tools', 'tool_names', 'input', 'agent_scratchpad'],
    template=REACT_TEMPLATE
)


class ManualReActAgent:
    def __init__(self, llm, tools, prompt, max_iterations=6):
        self.llm            = llm
        self.prompt         = prompt
        self.max_iterations = max_iterations
        self.tools          = {t.name: t for t in tools}
        self.tools_str      = '\n'.join(t.name + ': ' + t.description for t in tools)
        self.tool_names     = ', '.join(self.tools.keys())

    def _build_prompt(self, question, scratchpad):
        return self.prompt.format(
            tools=self.tools_str,
            tool_names=self.tool_names,
            input=question,
            agent_scratchpad=scratchpad
        )

    def _parse_action(self, text):
        action_match = re.search(r'Action:\s*(.+)', text)
        input_match  = re.search(r'Action Input:\s*(.+)', text, re.DOTALL)
        if action_match and input_match:
            return action_match.group(1).strip(), input_match.group(1).strip()
        return None, None

    def _parse_final_answer(self, text):
        match = re.search(r'Final Answer:\s*(.+)', text, re.DOTALL)
        return match.group(1).strip() if match else None

    def invoke(self, question):
        scratchpad = ''
        seen_calls = set()
        print('\n' + '='*65)
        print('QUESTION:', question[:100], '...')
        print('='*65)

        for iteration in range(1, self.max_iterations + 1):
            print(f'\n--- Iteration {iteration} ---')
            prompt_text = self._build_prompt(question, scratchpad)
            llm_output  = self.llm.invoke(prompt_text)

            if 'Observation:' in llm_output:
                llm_output = llm_output[:llm_output.index('Observation:')].rstrip()

            print(llm_output)

            final = self._parse_final_answer(llm_output)
            if final:
                print('\n' + '='*65)
                print('FINAL ANSWER:', final[:500])
                print('='*65)
                return final

            action, action_input = self._parse_action(llm_output)

            if action is None:
                print('[Agent] Could not parse Action.')
                scratchpad += llm_output + '\nObservation: Format error. Use JSON for Action Input.\n'
                continue

            call_key = (action, action_input.strip().lower()[:100])
            if call_key in seen_calls:
                print('[Agent] Duplicate call detected -- forcing Final Answer.')
                scratchpad += llm_output.rstrip() + '\nObservation: Already called this. Write Final Answer now.\nThought:'
                continue
            seen_calls.add(call_key)

            if action not in self.tools:
                observation = 'Tool not found: ' + action + '. Available: ' + self.tool_names
            else:
                print(f'\n[Calling MCP tool: {action}]')
                observation = str(self.tools[action].func(action_input))

            print('Observation:', observation[:400])
            scratchpad += llm_output.rstrip() + '\nObservation: ' + observation + '\nThought:'

        print('[Agent] Reached max_iterations.')
        return None


print('Infrastructure loaded.')
````
